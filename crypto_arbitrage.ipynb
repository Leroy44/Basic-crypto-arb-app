```python
1. For each of the three dates, measure the arbitrage spread between the two exchanges
```

```python
# measure the arbitrage spread between the two exchanges in three time slices: early, middle, late)

arbitrage_spread_early = abs(coinbase_sliced.loc["2018-01-29"] - bitstamp_sliced.loc["2018-01-29"])
arbitrage_spread_middle = abs(coinbase_sliced.loc["2018-02-06"] - bitstamp_sliced.loc["2018-02-06"])
arbitrage_spread_late = abs(coinbase_sliced.loc["2018-03-09"] - bitstamp_sliced.loc["2018-03-09"])

#...use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.
arbitrage_spread_early[arbitrage_spread_early > 0].describe()
```

```python
arbitrage_spread_middle[arbitrage_spread_middle > 0].describe()
```

```python
arbitrage_spread_late[arbitrage_spread_late > 0].describe()
```

2.  For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).

```python
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early = arbitrage_spread_early[arbitrage_spread_early > 0] / bitstamp_sliced.loc["2018-01-29"]
spread_return_early.tail(10)
```

```python
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0] / bitstamp_sliced.loc["2018-02-06"]
spread_return_middle.tail(10)
```

```python
spread_return_late = arbitrage_spread_late[arbitrage_spread_late > 0] / bitstamp_sliced.loc["2018-03-09"]
spread_return_late.tail(10)
```

4.  For the date early in the dataset, determine the number of times your trades with positive returns

```python
# exceed the 1% minimum threshold (.01) that you need to cover your costs
profitable_trades_early = spread_return_early[spread_return_early >= 0.01]
profitable_trades_early.head()
```

```python
profitable_trades_middle = spread_return_middle[spread_return_middle >= 0.01]
profitable_trades_middle.head()
```

```python
profitable_trades_late = spread_return_late[spread_return_late >= 0.01]
profitable_trades_late.head()
```

```python
profitable_trades_early.describe()
```

```python
profitable_trades_middle.describe()
```

```python
profitable_trades_late.describe()
```

5.  For the date early in the dataset, calculate the potential profit per trade in dollars 

```python
# Multiply the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early * bitstamp_sliced.loc["2018-01-29"]
profit_middle = profitable_trades_middle * bitstamp_sliced.loc["2018-02-06"]
profit_late = profitable_trades_late * bitstamp_sliced.loc["2018-03-09"]

# Drop any missing values from the profit DataFrame
profit_per_trade_early = profit_early.dropna()
profit_per_trade_middle = profit_middle.dropna()
profit_per_trade_late = profit_late.dropna()

# View the early profit DataFrame
profit_per_trade_early.head()
```
